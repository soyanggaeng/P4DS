```python
import os
import struct

def generate_random_float():
    # Generate 8 bytes of random data
    random_bytes = os.urandom(8)
    # Convert the bytes to a float between 0 and 1
    random_float = struct.unpack('d', random_bytes)[0]
    # Scale the random number to the desired range (0 to 5)
    scaled_random = random_float * 5
    return scaled_random

# Generate and print 10 random float numbers in the range of 0 to 5
for _ in range(10):
    random_float = generate_random_float()
    print(random_float)
```
